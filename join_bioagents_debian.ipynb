# parse the debian med metadata provided by Steffen in the local edam.json file

```python
import json
debian_med_metadata = json.load(open('edam.json'))
print(len(debian_med_metadata))
for entry in debian_med_metadata:
    print(entry['package'],entry['bio.agents'], entry['doi'])
bioagents_linked_packages = [entry for entry in debian_med_metadata if entry['bio.agents'] is not None]
print(len(bioagents_linked_packages))
print(len([entry['doi'] for entry in debian_med_metadata]))
```

# retrieve the bio.agents metadata and store them in a local bioagents.json file

```python
import requests
bioagents_list = []
next_bioagents_page = '?page=1'
print('retrieving the bio.agents data, please wait...')
while next_bioagents_page is not None:
    page = requests.get('https://bio.agents/api/agent/?format=json&' + next_bioagents_page[1:]).json()
    bioagents_list += page['list']
    next_bioagents_page = page.get('next',None)
json.dump(bioagents_list,open('bioagents.json','w'))
print('done!')
```

# create dataframes to store these data

```python
import pandas
from pandas.io.json import json_normalize
bioagents_df = json_normalize(bioagents_list)
# lowercase version of bio.agents ID in bio.agents dataframe
bioagents_df['bt_id_lc'] = bioagents_df['bioagentsID'].str.lower()
# lowercase version of debian ID in debian dataframe
debian_df = pandas.read_json('edam.json', orient='values')
debian_df['db_id_lc'] = debian_df['bio.agents'].str.lower()
merged_df = bioagents_df.merge(debian_df, how='right', left_on='bt_id_lc', right_on='db_id_lc', suffixes=('_bioagents', '_debian'))
print(merged_df.columns)
#merged_df = merged_df[['bio.agents','package','doi','publication']]
merged_df = merged_df[['bt_id_lc', 'db_id_lc', 'bio.agents','doi','publication','package','bioagentsID']]
print(merged_df[merged_df.db_id_lc.notnull()][merged_df.bioagentsID.isnull()])
print(merged_df[merged_df.db_id_lc.notnull()][merged_df.bioagentsID.isnull()].shape[0])
```

# print the number of entries in debian with a bio.agents ID and the number that have a corresponding bio.agents entry based on reference publications

```python
# find potential debian med entries which have not yet been mapped to bio.agents entries
debian_df_bydoi = debian_df.set_index('doi')
intersection_id = set()
intersection_doi = set()
debian_new_bioagents_links = []
entries = []
for idx, row in bioagents_df.iterrows():
    if row['bt_id_lc'] in debian_df['db_id_lc'].values:
        intersection_id.add(row['bt_id_lc'])
    else:
        for pub in row['publication']:
            if pub['doi'] and pub['type']=='Primary' and pub['doi'] in debian_df_bydoi.index and debian_df_bydoi.loc[pub['doi']]['db_id_lc'] is None:
                intersection_doi.add(row['bt_id_lc'])
                debian_new_bioagents_links.append({'bio.agents': row['bt_id_lc'], 'debian': debian_df_bydoi.loc[pub['doi']]['package']})
print(len(intersection_id), len(intersection_doi))
print('bio.agents\tdebian')
for entry in debian_new_bioagents_links:
    print(entry['bio.agents'] + "\t" + entry['debian'])
```
